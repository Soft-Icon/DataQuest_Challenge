```python
dataset_path = "/content/10ALYTICS_MONTHLY_KNOWLEDGEQUEST/e-commerce dataset/"
```

```python
#import necessary libraries
import pandas as pd
import os
```

```python
# List of all CSV files to load
csv_files = [
    'customers.eCommerce.csv',
    'geolocation.eCommerce.csv',
    'leads_closed.eCommerce.csv',
    'leads_qualified.eCommerce.csv',
    'order_items.eCommerce.csv',
    'order_payments.eCommerce.csv',
    'order_reviews.eCommerce.csv',
    'orders.eCommerce.csv',
    'product_category_name_translation.eCommerce.csv',
    'products.eCommerce.csv',
    'sellers.eCommerce.csv'
]
# prepare an empty dictionary for use as dataframe
dataframes = {}
```

```python
#Loading and inspecting datasets
for file in csv_files:
    file_path = os.path.join(dataset_path, file)
    df_name = file.replace('.eCommerce.csv', '')
    try:
        df = pd.read_csv(file_path)
        dataframes[df_name] = df
        print(f'\n--- {df_name} ---')
        print(f'Shape: {df.shape}')
        print(f'Columns: {df.columns.tolist()}')
        print('\nHead:')
        print(df.head())
        print('\nInfo:')
        df.info()
    except Exception as e:
        print(f'Error loading {file}: {e}')
```

## Data Cleaning

```python
#customers.eCommerce.csv
customers_df = dataframes["customers"]

#No obvious missing values or type issues from initial inspection, but then check for duplicates
customers_df.drop_duplicates(inplace=True)
print(f"customers_df cleaned. New shape: {customers_df.shape}")
```

```python
#geolocation.eCommerce.csv
geolocation_df = dataframes["geolocation"]

# Drop duplicates, especially important for location data
geolocation_df.drop_duplicates(inplace=True)
print(f"geolocation_df cleaned. New shape: {geolocation_df.shape}")
```

```python
#leads_closed.eCommerce.csv
leads_closed_df = dataframes["leads_closed"]

# The Correct column name for leads_closed_df is 'won_date'
if 'won_date' in leads_closed_df.columns:
    leads_closed_df["won_date"] = pd.to_datetime(leads_closed_df["won_date"])
else:
    print("Warning: 'won_date' column not found in leads_closed_df.")

leads_closed_df.drop_duplicates(inplace=True)
print(f"leads_closed_df cleaned. New shape: {leads_closed_df.shape}")
```

```python
#leads_qualified.eCommerce.csv
leads_qualified_df = dataframes["leads_qualified"]

# parse date correctly for column named 'first_contact_date'
leads_qualified_df["first_contact_date"] = pd.to_datetime(leads_qualified_df["first_contact_date"])
leads_qualified_df.drop_duplicates(inplace=True)
print(f"leads_qualified_df cleaned. New shape: {leads_qualified_df.shape}")
```

```python
# order_items.eCommerce.csv
order_items_df = dataframes["order_items"]

# Check for missing values and duplicates
order_items_df.dropna(inplace=True) # Drop rows with any missing values
order_items_df.drop_duplicates(inplace=True)
print(f"order_items_df cleaned. New shape: {order_items_df.shape}")
```

```python
# order_payments.eCommerce.csv
order_payments_df = dataframes["order_payments"]

# Check for missing values and duplicates
order_payments_df.dropna(inplace=True)
order_payments_df.drop_duplicates(inplace=True)
print(f"order_payments_df cleaned. New shape: {order_payments_df.shape}")
```

```python
# order_reviews.eCommerce.csv ---
order_reviews_df = dataframes["order_reviews"]

# Convert timestamp columns to datetime
order_reviews_df["review_creation_date"] = pd.to_datetime(order_reviews_df["review_creation_date"], errors='coerce')
order_reviews_df["review_answer_timestamp"] = pd.to_datetime(order_reviews_df["review_answer_timestamp"], errors='coerce')

# Fill missing review_comment_title and review_comment_message with empty string
order_reviews_df["review_comment_title"].fillna("", inplace=True)
order_reviews_df["review_comment_message"].fillna("", inplace=True)
order_reviews_df.dropna(subset=["review_score"], inplace=True) # Drop rows where review_score is missing
order_reviews_df.drop_duplicates(inplace=True)
print(f"order_reviews_df cleaned. New shape: {order_reviews_df.shape}")
```

```python
#orders.eCommerce.csv
orders_df = dataframes["orders"]

# Convert timestamp columns to datetime
for col in ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date",
            "order_delivered_customer_date", "order_estimated_delivery_date"]:
    orders_df[col] = pd.to_datetime(orders_df[col], errors='coerce') # Coerce errors will turn unparseable dates into NaT

# Drop rows where critical timestamps are missing (e.g., purchase timestamp)
orders_df.dropna(subset=["order_purchase_timestamp"], inplace=True)
orders_df.drop_duplicates(inplace=True)
print(f"orders_df cleaned. New shape: {orders_df.shape}")
```

```python
#product_category_name_translation.eCommerce.csv
product_category_translation_df = dataframes["product_category_name_translation"]
product_category_translation_df.drop_duplicates(inplace=True)
print(f"product_category_translation_df cleaned. New shape: {product_category_translation_df.shape}")
```

```python
# products.eCommerce.csv
products_df = dataframes["products"]

# Fill missing product_category_name with \'unknown\'
products_df["product_category_name"].fillna("unknown", inplace=True)

# Fill missing numerical product attributes with 0 or median, depending on context. assume, 0.
for col in ["product_name_lenght", "product_description_lenght", "product_photos_qty", "product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"]:
    products_df[col].fillna(0, inplace=True)
products_df.drop_duplicates(inplace=True)
print(f"products_df cleaned. New shape: {products_df.shape}")
```

```python
# sellers.eCommerce.csv
sellers_df = dataframes["sellers"]
sellers_df.drop_duplicates(inplace=True)
print(f"sellers_df cleaned. New shape: {sellers_df.shape}")

# Save cleaned dataframes to a new dictionary
cleaned_dataframes = {
    "customers": customers_df,
    "geolocation": geolocation_df,
    "leads_closed": leads_closed_df,
    "leads_qualified": leads_qualified_df,
    "order_items": order_items_df,
    "order_payments": order_payments_df,
    "order_reviews": order_reviews_df,
    "orders": orders_df,
    "product_category_name_translation": product_category_translation_df,
    "products": products_df,
    "sellers": sellers_df
}
```

## Data Integration and Feature Engineering

*Integration for customer insights*

```python
# Merge orders with customers
customers_df2 = cleaned_dataframes['customers']
orders_df2 = cleaned_dataframes['orders']
customers_orders_df = pd.merge(customers_df2, orders_df2, on='customer_id', how='left')

customers_orders_df.head(5)
```

```python
#Merge with order_items and products for product details
order_items_df2 = cleaned_dataframes['order_items']
products_df2 = cleaned_dataframes['products']
product_category_translation_df2 = cleaned_dataframes['product_category_name_translation']
customers_orders_items_products_df = pd.merge(customers_orders_df, order_items_df2, on='order_id', how='left')
customers_orders_items_products_df = pd.merge(customers_orders_items_products_df, products_df2, on='product_id', how='left')
customers_orders_items_products_df = pd.merge(customers_orders_items_products_df, product_category_translation_df2, on='product_category_name', how='left')

# Merge with geolocation for customer city/state
# Note: geolocation_zip_code_prefix might not be unique, so handle carefully (e.g., take first match)
geolocation_df2 = cleaned_dataframes['geolocation']
geolocation_unique_df = geolocation_df2.drop_duplicates(subset=['geolocation_zip_code_prefix'])
customers_orders_items_products_geo_df = pd.merge(customers_orders_items_products_df, geolocation_unique_df, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left')

# Feature Engineering for Customer Insights
# Total purchase value per order item
customers_orders_items_products_geo_df['item_total_price'] = customers_orders_items_products_geo_df['price'] + customers_orders_items_products_geo_df['freight_value']
```

*Integration for Sales Performance Analysis*

```python
# Merge orders with order_payments
order_payments_df = cleaned_dataframes['order_payments']
orders_payments_df = pd.merge(orders_df2, order_payments_df, on='order_id', how='left')

# Merge order_items with sellers for seller performance
sellers_df = cleaned_dataframes['sellers']
order_items_sellers_df = pd.merge(order_items_df2, sellers_df, on='seller_id', how='left')

# Merge leads data
leads_qualified_df2 = cleaned_dataframes['leads_qualified']
leads_closed_df2 = cleaned_dataframes['leads_closed']
leads_combined_df = pd.merge(leads_qualified_df2, leads_closed_df2, on='mql_id', how='left')
```

*Integration for Delivery & Logistics*

```python
# Delivery duration and variance (already handled in orders_df cleaning for datetime conversion)
orders_df2['delivery_duration'] = (orders_df2['order_delivered_customer_date'] - orders_df2['order_purchase_timestamp']).dt.days
orders_df2['delivery_variance'] = (orders_df2['order_delivered_customer_date'] - orders_df2['order_estimated_delivery_date']).dt.days
```

```python
orders_df2.head(3)
```

*Integration for Review & Satisfaction Insights*

```python
# Merge order_reviews with orders and order_items
order_reviews_df = cleaned_dataframes['order_reviews']
order_reviews_orders_df = pd.merge(order_reviews_df, orders_df2, on='order_id', how='left')
order_reviews_orders_items_df = pd.merge(order_reviews_orders_df, order_items_df2, on='order_id', how='left')
order_reviews_orders_items_products_df = pd.merge(order_reviews_orders_items_df, products_df2, on='product_id', how='left')
order_reviews_orders_items_products_df = pd.merge(order_reviews_orders_items_products_df, product_category_translation_df2, on='product_category_name', how='left')
order_reviews_orders_items_products_sellers_df = pd.merge(order_reviews_orders_items_products_df, sellers_df, on='seller_id', how='left')
```

## Save all processed csv for further analysis

```python
output_dir = "/content/drive/MyDrive/processed_data/"
os.makedirs(output_dir, exist_ok=True)

customers_orders_items_products_geo_df.to_csv(os.path.join(output_dir, "processed_customer_product_geo_data.csv"), index=False)
orders_payments_df.to_csv(os.path.join(output_dir, "processed_orders_payments_data.csv"), index=False)
leads_combined_df.to_csv(os.path.join(output_dir, "processed_leads_data.csv"), index=False)
orders_df.to_csv(os.path.join(output_dir, "processed_orders_delivery_data.csv"), index=False) # orders_df now has delivery features
order_reviews_orders_items_products_sellers_df.to_csv(os.path.join(output_dir, "processed_reviews_data.csv"), index=False)
```
